```python
#Method 2:
#Checking total number of missing value in a column
mydata.isnull().sum()
```

**ERROR ALER!!**
* Notice how the result are all zero (0) eventhough we are sure there is a missing value in our data
* The missing data in our data appeared as question mark (?)
* So question mark (?) in python is considered as a string
* So numpy library read the data string as object

**Replace the missing value '?' to NaN numpy**

```python
#replace the '?' by NaN numpy
import numpy as np

mydata.replace('?', np.nan, inplace = True)

#Checking total number of missing value in a column AGAIN
mydata.isnull().sum()
```

## Dealing With Missing Value In "Normalized-Losses"

```python
mean_norm = mydata['normalized-losses'].mean()

mean_norm
```

**ERROR ALERT!!**

* Remember that the 'normalized-loses' is an "**object**" so **CONVERT** it to "**float**"

**DANGER!!!**
* If you convert float to integer, you will lose some data because it will round up the decimal number to a whole number
* But if you convert interger to float then it will not matter

```python
mean_norm = mydata['normalized-losses'].astype(float).mean()

mean_norm
```

**Replace Function**
* replace(value nak replace, value baru, inplace = True)

```python
mydata['normalized-losses'].replace(np.nan, mean_norm, inplace = True)

#Checking total number of missing value in a column AGAIN
mydata.isnull().sum()
```

## Dealing With Missing Value In "num-of-doors"

### 3 Methods To Check Frequency of A Columns

**Method 1: Using Describe Function**

```python
#Cange number of doors missing value with the most frequency categorical value

#check frequency of the categorical value on number of doors column
mydata.describe(include='all')
```

**Method 2: Using Count Function**

```python
mydata['num-of-doors'].value_counts()
```

**Method 3: Using Count Max Value Function**

```python
mydata['num-of-doors'].value_counts().idxmax()
```

* replace the "num-of-doors" missing value with 'four' the most frequent value in that column

```python
#Cara 1:
mydata['num-of-doors'].replace(np.nan,mydata['num-of-doors'].value_counts().idxmax(), inplace=True)

#Cara 2:
#mydata['num-of-doors'].replace(np.nan,'four', inplace=True)
#cara 2 pon betul (mana-mana pon boleh)

mydata.isnull().sum()
```

## Dealing With Missing Value In "bore"

```python
mean_bore= mydata['bore'].astype(float).mean()

mean_bore
```

```python
mydata['bore'].replace(np.nan, mean_bore, inplace = True)

#Checking total number of missing value the bore column AGAIN
mydata.isnull().sum()
```

## Dealing With The Rest of The Missing Value

```python
# Finding the mean value of the 'stroke' column
mean_stroke= mydata['stroke'].astype(float).mean()
mean_horsepower= mydata['horsepower'].astype(float).mean()
mean_peak= mydata['peak-rpm'].astype(float).mean()

# Replace the missing value in 'stroke' with the mean value
mydata['stroke'].replace(np.nan, mean_stroke, inplace = True)
mydata['horsepower'].replace(np.nan, mean_horsepower, inplace = True)
mydata['peak-rpm'].replace(np.nan, mean_peak, inplace = True)

#Checking total number of missing value the stroke column AGAIN
mydata.isnull().sum()
```

## Dealing With Missing Value In "Price"
* Here we are going to drop the missing value in price

```python
mydata.dropna(subset=['price'],axis=0, inplace=True)

#Checking total number of missing value the price column AGAIN
mydata.isnull().sum()
```
