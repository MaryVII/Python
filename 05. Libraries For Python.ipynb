**iloc Function** </p>



**iloc:** This is an integer-based indexing method in Pandas. It is used for selecting data by position or integer index. It allows you to select data by row numbers and column numbers (integer position). For example, df.iloc[row_index, column_index] selects data at the specified row and column indices.

```python
#df.iloc[row index,column index]

df_NFLX.iloc[2,3]
```

```python
df_NFLX.iloc[0:5,2:4]
```

**Loc Function - will only give output of  index of the row and name of column**

**loc:** This is a label-based indexing method in Pandas. It is used to access a group of rows and columns by labels or a boolean array. It allows you to select data by row labels and column labels. For example, df.loc[row_label, column_label] selects data at the specified row and column labels.

```python
#df.loc[[row index],[column name]]
df_NFLX.loc[[2],['High']]
```

### Section 10.9: Rename Column


**The inplace=True** parameter in Python, particularly in Pandas DataFrame operations, specifies whether to modify the DataFrame directly or return a new modified copy of the DataFrame. Heres how it works:

**inplace=True**
* When inplace=True is used in a Pandas operation (e.g., .drop() or .fillna()), it modifies the DataFrame directly.
* This means that the changes are applied to the original DataFrame itself, and the operation does not return a new DataFrame.

**inplace=False**
* When inplace=False or if inplace is not specified (which defaults to False), the operation returns a new modified DataFrame without affecting the original DataFrame.
* This allows you to chain operations or keep the original DataFrame intact.

```python
df_NFLX.rename(columns = {'High': 'Test'}, inplace = True) #Kalau kita letak 'inplace = true' dia akan kacau original data
df_NFLX.head()
```

```python
df_NFLX_test = df_NFLX.rename(columns = {'Low': 'Test'})
df_NFLX_test.head()
```
