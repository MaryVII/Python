### Difference Between np.linspace and np.random function

* **numpy.linspace (np.linspace)**
  * Purpose: To generate evenly spaced numbers over a specified range.
  * Usage: numpy.linspace(start, stop, num)

  * Parameters:
    * start: The starting value of the sequence.
    * stop: The end value of the sequence.
    * num: The number of samples to generate (including the start and stop values).

* **numpy.random (np.random)**
  * Purpose: To generate random numbers.
  * Usage: numpy.random provides several methods to generate random numbers, including:
    * numpy.random.rand(d0, d1, ..., dn): Uniform distribution over [0, 1).
    * numpy.random.randn(d0, d1, ..., dn): Standard normal distribution (mean 0, variance 1).
    * numpy.random.randint(low, high=None, size=None, dtype='l'): Random integers from low (inclusive) to high (exclusive).
    * numpy.random.normal(loc=0.0, scale=1.0, size=None): Normal (Gaussian) distribution.
And many others.

**Summary**
* numpy.linspace: Generates a specified number of evenly spaced values over a given range. Useful for deterministic sequences where the exact values and their spacing are important.
* numpy.random: Generates random numbers from various distributions. Useful for creating random samples, simulations, or any situation where randomness is required.

**(mean,std,number of samples)**

**Refer code below: In the code "col1 = np.random.normal(250, 10, 500)", the parameters 250, 10, and 500 represent the following:**

* 250: This is the **mean (average)** of the normal distribution. It determines the center of the distribution.

* 10: This is the **standard deviation** of the normal distribution. It measures the spread or dispersion of the data around the mean. A smaller standard deviation means the data points are closer to the mean, while a larger standard deviation means they are more spread out.

* 500: This is the **number of samples** to be generated from the normal distribution. It specifies how many data points will be created.

In summary, np.random.normal(250, 10, 500) generates 500 random samples from a normal distribution with a mean of 250 and a standard deviation of 10.

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

#create numerical data (or numerical column)
#all those column that will be int or float

col1 = np.random.normal(250, 10, 500) #(mean,std,number of samples)
col2 = np.random.normal(100, 20, 500)
col3 = np.random.normal(750, 5, 500)

#create categorical data

categorical_data = ['A','B','C','D','F','G','H','I','J']

col4 = np.random.choice(categorical_data, 500)
col5 = np.random.choice(categorical_data, 500) #use np.random.choice for categorical data

#create data time

start_date = datetime(2022, 1, 1)
end_date = datetime(2024, 1, 1)

delta = end_date - start_date

date_list = [start_date + timedelta(days=x) for x in range(delta.days)]

col6 = np.random.choice(date_list, 500)

#combine all the columns

data = np.column_stack((col1,col2,col3,col4,col5,col6))

df = pd.DataFrame(data)
df.head()
```

Adding "k- -" will change the colour of the line to black

```python
x = [1,2,3,4,5,6,7,8]
y = [4,7,9,10,13,15,17,19]
plt.plot(x,y,'k--')

x1 = [2,3,5,6,7,8,10,11]
y1 = [3,5,8,11,14,16,18,19]
plt.plot(x1,y1,'--')

plt.xlabel('Number of Years')
plt.ylabel('Salary in RM')
plt.title('Salary in Kuala Lumpur')
plt.legend(['Company A','Company B'])

plt.show()
```

**plt.plot(x, y, 'k--'):**

Here, 'k--' is a format string.
* 'k' specifies the color black (k stands for "black").
* '--' specifies a dashed line style.
* So, this command plots the data with a black dashed line.
plt.plot(x, y, '--'):

Here, only '--' is provided, which specifies a dashed line style.
* No color is explicitly specified.
* When the color is not specified, matplotlib uses the default color cycle to automatically assign a color to the plot.
* This cycle typically includes a set of different colors to distinguish multiple plots.
