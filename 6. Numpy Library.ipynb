### Section 11: Create a sample dataset with various types of data (numerical, categorical, and datetime)

**(mean,std,number of samples)**

**In the code col1 = np.random.normal(250, 10, 500), the parameters 250, 10, and 500 represent the following:**

* 250: This is the **mean (average)** of the normal distribution. It determines the center of the distribution.

* 10: This is the **standard deviation** of the normal distribution. It measures the spread or dispersion of the data around the mean. A smaller standard deviation means the data points are closer to the mean, while a larger standard deviation means they are more spread out.

* 500: This is the **number of samples** to be generated from the normal distribution. It specifies how many data points will be created.

In summary, np.random.normal(250, 10, 500) generates 500 random samples from a normal distribution with a mean of 250 and a standard deviation of 10.

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

#create numerical data (or numerical column)
#all those column that will be int or float

col1 = np.random.normal(250, 10, 500) #(mean,std,number of samples)
col2 = np.random.normal(100, 20, 500)
col3 = np.random.normal(750, 5, 500)

#create categorical data

categorical_data = ['A','B','C','D','F','G','H','I','J']

col4 = np.random.choice(categorical_data, 500)
col5 = np.random.choice(categorical_data, 500)

#create data time

start_date = datetime(2022, 1, 1)
end_date = datetime(2024, 1, 1)

delta = end_date - start_date

date_list = [start_date + timedelta(days=x) for x in range(delta.days)]

col6 = np.random.choice(date_list, 500)

#combine all the columns

data = np.column_stack((col1,col2,col3,col4,col5,col6))

df = pd.DataFrame(data)
df.head()
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

#create numerical data (or numerical column)
#all those column that will be int or float

col1 = np.random.normal(250, 10, 500)
col2 = np.random.normal(100, 20, 500)
col3 = np.random.normal(750, 5, 500)

#create categorical data

categorical_data = ['A','B','C','D','F','G','H','I','J']

col4 = np.random.choice(categorical_data, 500)
col5 = np.random.choice(categorical_data, 500)

#create data time

start_data = datetime(2022, 1, 1) # This variable was originally named 'start_data'
end_date = datetime(2024, 1, 1)

# Use 'start_data' instead of 'start_date'
delta = end_date - start_data

date_list = [start_data + timedelta(days=x) for x in range(delta.days)]

col6 = np.random.choice(date_list, 500)

#combine all the columns

data = np.column_stack((col1,col2,col3,col4,col5,col6))

df = pd.DataFrame(data, columns =['Num1','Num2','Num3','Cat1','Cat2','Date'])
df.head() # Call the head() method to display the DataFrame
```

Converting object datatype to float/interger
