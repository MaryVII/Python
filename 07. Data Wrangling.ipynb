## Dealing With Missing Value In "Normalized-Losses"

```python
mean_norm = mydata['normalized-losses'].mean()

mean_norm
```

**ERROR ALERT!!**

* Remember that the 'normalized-loses' is an "**object**" so **CONVERT** it to "**float**"

**DANGER!!!**
* If you convert float to integer, you will lose some data because it will round up the decimal number to a whole number
* But if you convert interger to float then it will not matter

```python
mean_norm = mydata['normalized-losses'].astype(float).mean()

mean_norm
```

**Replace Function**
* replace(value nak replace, value baru, inplace = True)

```python
mydata['normalized-losses'].replace(np.nan, mean_norm, inplace = True)

#Checking total number of missing value in a column AGAIN
mydata.isnull().sum()
```

## Data Normalization

```python
#Checking the width data
mydata['width'].head()
```

```python
#Checking the width, length, height data
mydata[['width','length','height']].head()
```

### Data Normalization: Using Simple Feature Scaling

```python
#Simple Feature Scaling for the width column
mydata['width'] = mydata['width']/mydata['width'].max()
mydata[['width','length','height']].head()
```

```python
mydata['width'].describe()
```

### Data Normalization: Using Min-Max Method

```python
#min-max normalization for length column
mydata['length'] = (mydata['length']-mydata['length'].min())/(mydata['length'].max()-mydata['length'].min())
mydata[['width','length','height']].head()
```

### Data Normalization: Using Z-Score Method

```python
#z-score normalization for height column
mydata['height'] = (mydata['height']-mydata['height'].mean())/mydata['height'].std()
mydata[['width','length','height']].head()
```

```python
#Now let's compare all the data normalization
#Usually we don't use all three methods at once
#It is recommended to only use one data normalization method throughout the project
#People usually use the z-score method
mydata[['width','length','height']].describe()
```

## Data Binning

* **By performing these steps, you:**

  * **Standardize Your Data:** Convert continuous variables into categorical variables, which can be useful for certain types of analysis and modeling.
  * **Enhance Machine Learning Models:** Use the categorized data as features in machine learning models, potentially improving model performance by highlighting meaningful intervals.
  * **Improve Data Visualization:** Make data visualization more intuitive by grouping continuous data into labeled categories, which can simplify the interpretation of charts and graphs.

This process of binning and labeling continuous data is a valuable technique in data preprocessing, enabling more effective data analysis and feature engineering.

```python
#Checking the min and max value of the 'price'
mydata['price'].describe()
```

```python
#Creating Bins for 'price'
price_bins = np.linspace(mydata_new1['price'].min(), mydata_new1['price'].max(), num=4)

#Printing the bins
print(price_bins)

#Label the bins
price_groups = ['Low Price','Mid Price','High Price']

#Assigning Data to the bin
mydata_new1['price_bins'] = pd.cut(mydata_new1['price'],price_bins, labels = price_groups, include_lowest=True)

#Viewing the Binned Data
mydata_new1[['price','price_bins']].head(10)
```

```python
#The count category in the new price bin
mydata_new1['price_bins'].value_counts()
```

```python
#checking the overall data
mydata_new1.describe()
```

```python
#Creating Bins for 'normalized-losses'
norm_bins = np.linspace(mydata_new1['normalized-losses'].min(), mydata_new1['normalized-losses'].max(), num=4)

#Printing the bins
print(norm_bins)

#Label the bins
norm_groups = ['Low','Mid','High']

#Assigning Data to the bin
mydata_new1['normalized-losses_bins'] = pd.cut(mydata_new1['normalized-losses'],norm_bins, labels = norm_groups, include_lowest=True)

#Viewing the Binned Data
mydata_new1[['normalized-losses','normalized-losses_bins']].head(10)
```

```python
#The count category in the new norm bin
mydata_new1['normalized-losses_bins'].value_counts()
```

**Summary from above code**

* **What Are You Doing?**
  * You are categorizing the continuous data in the normalized-losses column into discrete bins labeled 'Low', 'Mid', and 'High'.
  * This process is known as binning, which helps in transforming continuous numerical data into categorical data based on defined ranges.

* **What Do You Learn?**
  * Creating Bins: How to create equally spaced intervals (bins) using numpy.linspace.
  * Labeling Bins: How to assign meaningful labels to these intervals.
  * Using pandas.cut: How to categorize (bin) continuous data into these labeled bins.
  * Inspecting Binned Data: How to verify the binning process by inspecting the results.
* **Why Is This Important?**
  * Simplifies Analysis: Binning converts continuous data into categorical data, making it easier to analyze and interpret.
  * Feature Engineering: Binned data can be used as a feature in machine learning models to improve their performance.
  * Visualization: Categorical data is often easier to visualize and understand in charts and graphs.
